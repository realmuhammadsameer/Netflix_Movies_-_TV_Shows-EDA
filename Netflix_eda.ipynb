```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Netflix_Movies_and_TV_Shows_cleaned.csv')
```

```python
df.info()
```

# Observation:

- The dataset contains 8797 rows and 12 columns.
- Most columns have appropriate data types.
- Few/no missing values are present.

```python
df.describe()
```

# Observation:

- The average release year is around 2014.
- Most content was released between 2013 and 2019.
- The dataset contains titles ranging from 1925 to 2021.

```python
df['type'].value_counts()
```

# Observation:

- Movies are more common than TV Shows.
- Some categories appear much more frequently than others.

```python
df['release_year'].hist()
plt.show()
```

# Observation:

- Most content was released in recent years.
- The distribution is concentrated after 2010.

```python
sns.boxplot(x=df['release_year'])
plt.show()
```

# Observation:

- A few outliers are visible in the dataset.
- Most values lie within a limited range.

```python
sns.scatterplot(x='release_year', y='duration', data=df)
plt.show()
```

# Observation: 

- No strong relationship is observed between the selected variables.
- The points are widely scattered.

```python
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True)
plt.show()
```

# Observation:

- Most numerical variables have weak correlations.
- No strong positive or negative correlation is present.

```python
sns.pairplot(df.select_dtypes(include='number'))
plt.show()
```

# Observation:

- The relationships among numerical features appear weak.
- No clear clusters are visible.

# Summary of Findings

1. The dataset contains mostly recent content.
2. Movies are more common than TV Shows.
3. Numerical features show weak correlations.
4. A few outliers are present in the data.
5. No strong relationships were identified between the selected variables.
6. Overall, the dataset is suitable for exploratory analysis and visualization.
